# Stratégie à adopter

Imputer les données. L'imputation permet de conserver toutes les observations en remplaçant les valeurs manquantes par des estimations raisonnables. Il peut aider à réduire ce biais en fournissant des estimations pour les valeurs manquantes.

# Imputez les valeurs manquantes en utilisant la médiane pour chaque colonne.


```python
variables_suspectes = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[variables_suspectes] = df[variables_suspectes].replace(0, np.nan)

for var in variables_suspectes:
    median_value = df[var].median()
    df[var].fillna(median_value, inplace=True)

print(df.isnull().sum())
```

# Les différences observées dans les graphiques sont-elles statistiquement significatives ?

```python

from scipy.stats import ttest_ind

df = pd.read_csv('/Users/elliot/Desktop/diabetes.csv')

variables_suspectes = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[variables_suspectes] = df[variables_suspectes].replace(0, np.nan)

for var in variables_suspectes:
    median_value = df[var].median()
    df[var].fillna(median_value, inplace=True)

group_0 = df[df['Outcome'] == 0]
group_1 = df[df['Outcome'] == 1]

t_stat_glucose, p_value_glucose = ttest_ind(group_0['Glucose'], group_1['Glucose'])
t_stat_bmi, p_value_bmi = ttest_ind(group_0['BMI'], group_1['BMI'])

print(f"Test t de Student pour le Glucose : t-stat = {t_stat_glucose}, p-value = {p_value_glucose}")
print(f"Test t de Student pour le BMI : t-stat = {t_stat_bmi}, p-value = {p_value_bmi}")
```

Les différences observées dans les boxplots pour le glucose et le BMI sont donc statistiquement significatives, ce qui signifie qu'il est très peu probable que ces différences soient dues au hasard. Ces résultats renforcent l'idée que le glucose et le BMI sont des indicateurs importants du diabète.

# Identifier les patientes à haut risque.

```python
seuil_glucose = 140
seuil_bmi = 30
seuil_age = 50

patientes_haut_risque = df[(df['Glucose'] > seuil_glucose) & (df['BMI'] > seuil_bmi) & (df['Age'] > seuil_age)]

print("Patientes à haut risque :")
print(patientes_haut_risque)
```

Les patientes identifiées comme étant à haut risque sont celles qui ont des niveaux de glucose supérieurs à 140 mg/dL, un BMI supérieur à 30 et un âge supérieur à 50 ans. Ces patientes sont plus susceptibles de développer des complications liées au diabète et devraient être surveillées de près.

# Un modèle simple de seuil basé sur les trois variables (Glucose, BMI, Age) pour prédire si une patiente a du diabète.

```python
# Définir les seuils
seuil_glucose = 140
seuil_bmi = 30
seuil_age = 50

def predire_diabete(row):
    if row['Glucose'] > seuil_glucose and row['BMI'] > seuil_bmi and row['Age'] > seuil_age:
        return 1
    else:
        return 0

df['Prediction'] = df.apply(predire_diabete, axis=1)

print(df[['Glucose', 'BMI', 'Age', 'Outcome', 'Prediction']].head(10))

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

accuracy = accuracy_score(df['Outcome'], df['Prediction'])
print(f"Accuracy: {accuracy}")

conf_matrix = confusion_matrix(df['Outcome'], df['Prediction'])
print("Matrice de confusion :")
print(conf_matrix)

class_report = classification_report(df['Outcome'], df['Prediction'])
print("Rapport de classification :")
print(class_report)
```

# Les trois principales variables qui semblent influencer le plus le développement du diabète

Les variables Glucose, BMI et Age sont les trois principales variables qui semblent influencer le plus le développement du diabète. Ces variables montrent des différences significatives entre les patientes diabétiques et non diabétiques et sont des indicateurs clés du risque de diabète.
